# Adicionando Colunas, Modificando Colunas e Valores

### Vamos pegar o nosso dataframe novamente

```python
import pandas as pd
#importando os arquivos

pasta_referencia = r'.\Pandas 05 - Dados'
vendas_df    = pd.read_csv(fr'{pasta_referencia}\Contoso - Vendas - 2017.csv'    , sep=';', encoding='utf-8')
produtos_df  = pd.read_csv(fr'{pasta_referencia}\Contoso - Cadastro Produtos.csv', sep=';', encoding='ANSI')
lojas_df     = pd.read_csv(fr'{pasta_referencia}\Contoso - Lojas.csv'            , sep=';', encoding='ANSI')
clientes_df  = pd.read_csv(fr'{pasta_referencia}\Contoso - Clientes.csv'         , sep=';', encoding='ANSI')
promocoes_df = pd.read_csv(fr'{pasta_referencia}\Contoso - Promocoes.csv'        , sep=';', encoding='ANSI')

#limpando apenas as colunas que queremos
clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto', 'Nome do Produto']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]

#mesclando e renomeando os dataframes
vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente').rename(columns={'E-mail': 'E-mail do Cliente'})
display(vendas_df)
```

```python
vendas_df.info()
```

### Agora, e se quisermos acrescentar uma coluna com o mÃªs, o dia e o ano de cada venda (e nÃ£o sÃ³ a data completa)

```python
vendas_df['Data da Venda'] = pd.to_datetime(vendas_df['Data da Venda'], format='%d/%m/%Y')
vendas_df['Ano da Venda']  = vendas_df['Data da Venda'].dt.year
vendas_df['Mes da Venda']  = vendas_df['Data da Venda'].dt.month
vendas_df['Dia da Venda']  = vendas_df['Data da Venda'].dt.day

vendas_df['Data do Envio'] = pd.to_datetime(vendas_df['Data do Envio'], format='%d/%m/%Y')
vendas_df['Ano da Envio']  = vendas_df['Data do Envio'].dt.year
vendas_df['Mes da Envio']  = vendas_df['Data do Envio'].dt.month
vendas_df['Dia da Envio']  = vendas_df['Data do Envio'].dt.day

vendas_df  = vendas_df.drop(columns=['Data da Venda', 'Data do Envio'])
display(vendas_df)
vendas_df.info()
```
