### Puxar a Base de Dados

```python
import pandas as pd

CAMINHO_REFERENCIA = r'.\Pandas 26 - Dados'
tabela = pd.read_csv(fr'{CAMINHO_REFERENCIA}\exportacoes_franca.csv')
tabela = tabela[tabela['Country'] == 'France']
display(tabela)
```

### Informações Gerais

Quais linhas têm a tabela? Existe algum valor vazio que deveria ser tratado?

```python
tabela.info()
```

### Como foi a evolução das exportações para a França ao longo dos anos?
Crie uma tabela com os anos de 2016, 2017, 2018, 2019 e 2020 e o valor US$ de exportações

```python
evolucao_exportacao_franca = tabela[['Year', 'US$ FOB']].groupby('Year').sum(numeric_only=True)

def formatar(valor: int) -> str:
    return f'US$ {valor:_.2f}'.replace(',', '.').replace('_', '.')

evolucao_exportacao_franca.plot(kind='bar', figsize=(15, 5))
evolucao_exportacao_franca['US$ FOB'] = evolucao_exportacao_franca['US$ FOB'].apply(formatar)
display(evolucao_exportacao_franca)



```

### Quais os produtos mais exportados ao logo de todo o período?
Crie uma tabela com os anos de 2016, 2017, 2018, 2019 e 2020 e o valor US$ de exportações

```python
produto = 'SH4 Description'
produtos_mais_exportados = tabela[['US$ FOB', produto]].groupby(produto).sum().sort_values(by='US$ FOB', ascending=False)
produtos_mais_exportados.head().plot(kind='bar', figsize=(15, 5))
produtos_mais_exportados['US$ FOB'] = produtos_mais_exportados['US$ FOB'].apply(formatar)

display(produtos_mais_exportados.head())
```

### Em 2020, qual cidade mais exportou para a França?
Em valors US$

```python
# cidade_mais_exportou_franca_2020 = tabela.loc[tabela['Year'] == 2020, :] # <- Outra opção de filtro do ano
ANO = 2020
cidade_mais_exportou_franca_2020 = tabela.loc[tabela['Year'] == ANO, :][['City', 'US$ FOB']].groupby('City', as_index=False) \
    .sum().sort_values(by='US$ FOB', ascending=False)
cidade_mais_exportou_franca_2020.head().plot(kind='bar', figsize=(15, 5))
cidade_mais_exportou_franca_2020['US$ FOB'] = cidade_mais_exportou_franca_2020['US$ FOB'].apply(formatar)

display(cidade_mais_exportou_franca_2020[:1])
```


### Quais os produtos mais exportados (em US$) que as 2 maiores cidades (em exportação em US$) mais exportaram?
Considerar apenas o ano de 2020

```python
PRODUTO = 'SH4 Description'
ANO = 2020
QTD_CIDADES = 2
tabela_Ano = tabela.loc[tabela['Year'] == ANO, :]
maiores_cidades_exportacao = list(tabela_Ano[['City', 'US$ FOB']].groupby('City', as_index=False).sum().sort_values('US$ FOB', ascending=False)['City'])

for cidade in maiores_cidades_exportacao[:QTD_CIDADES]:
    produtos_mais_exportados = tabela_Ano.loc[tabela_Ano['City'] == cidade, :][[PRODUTO, 'US$ FOB']].groupby(PRODUTO).sum(numeric_only=True).sort_values('US$ FOB', ascending=False).head()
    produtos_mais_exportados.plot(kind='bar', figsize=(15, 5), title=cidade, subplots=True)
    produtos_mais_exportados['US$ FOB'] = produtos_mais_exportados['US$ FOB'].apply(formatar)
    print(cidade)
    display(produtos_mais_exportados)
    print('-'*160)
```
