#### 2. Faça um Programa que peça um número (input) e então mostre a mensagem: "O número informado foi [número]."

```python
numero = input("Digite um número: ")

try:
    numero = float(numero)
    if numero.is_integer():
        numero = int(numero)
    print(f"O número digitado foi {numero}")
except:
    print(f"O termo digitado '{numero}' não é válido")

```

#### 3. Faça um Programa que peça dois números e imprima a soma.

```python
numerador = input("Digite o primeiro número: ")
denominador = input("Digite o segundo número: ")

try:
    numerador = float(numerador)
    denominador = float(denominador)
    if numerador.is_integer():
        numerador = int(numerador)
    if denominador.is_integer():
        denominador = int(denominador)
    print(f"A soma dos números {numerador} e {denominador} é igual à {numerador + denominador}")
except:
    print(f"O termo digitado '{numerador}' ou '{denominador}' não é válido")
```

#### 4. Faça um Programa que peça as 4 notas bimestrais de um aluno e mostre a média de todas as notas.

```python
totNotas = 4
media = 0
i = 0
while i < totNotas:
    numero = input(f"Digite a {i + 1}º nota: ")
    try:
        numero = float(numero)
        if numero < 0 or numero > 10:
            print(f"O número digitado '{numero}' não é válido")
            continue
        media += numero
        i += 1
    except:
        print(f"O termo digitado '{numero}' não é válido")
        continue

print(f"A média das notas digitadas é igual à {media/totNotas:.2f}")
```

#### 11. Tendo como dado de entrada a altura (h) de uma pessoa, construa um algoritmo que calcule seu peso ideal, utilizando as seguintes fórmulas:
##### a. Para homens: $P = 72,7h - 58$
##### b. Para mulheres: $P = 62,1h - 44,7$

```python
altura = float(input("Digite a altura da pessoa (em Metros): "))
genero = input("Digite o gênero da pessoa (M/F): ").upper()

if genero == "M":
    print(f"O peso ideal para a pessoa é de \033[1;36m{(72.7 * altura) - 58:,.2f}\033[0;40m Kg")
elif genero == "F":
    print(f"O peso ideal para a pessoa é de \033[1;36m{(62.1 * altura) - 44.7:,.2f}\033[0;40m Kg")
else:
    print(f"O gênero digitado '{genero}' não é válido")
```

#### 12. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
ganhoHora = float(input("Digite o valor ganho por hora: R$"))
horasTrabalhadas = float(input("Digite a quantidade de horas trabalhadas: R$"))
```

#####  Calcule o salário bruto (horas * salario por hora)

```python
salarioBruto = ganhoHora * horasTrabalhadas
print(f"O valor a ser pago é de R$\033[1;35m{salarioBruto:,.2f}\033[0;40m")
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
impostoRenda = salarioBruto * 0.11
print(f"O valor do imposto de renda é de R$\033[1;35m{impostoRenda:,.2f}\033[0;40m")
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
inss = salarioBruto * 0.08
print(f"O valor do INSS é de R$\033[1;35m{inss:,.2f}\033[0;40m")
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
sindicato = salarioBruto * 0.05
print(f"O valor do sindicato é de R$\033[1;35m{sindicato:,.2f}\033[0;40m")
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
salarioLiquido = salarioBruto - impostoRenda - inss - sindicato
print(f"O valor a ser pago é de R$\033[1;35m{salarioLiquido:,.2f}\033[0;40m")
```

#### 13. Faça um programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 3 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R\$ 80,00. Informe ao usuário a quantidades de latas de tinta a serem compradas e o preço total. (para simplificação nesse momento, não se preocupe em arredondar a quantidade de latas a serem compradas - vamos trabalhar isso em breve)

```python
from math import ceil
area = float(input("Digite a área a ser pintada (em m²): "))
rendimentoLata = 3
litrosUtilizados = area / rendimentoLata
volumeLata = 18
latas = ceil(litrosUtilizados / volumeLata)
precoLata = 80
custo = latas * precoLata
print(f"Em uma área de \033[1;32m{area:,.2f}\033[0;40m m²: ")
print(f"Serão necessárias \033[1;33m{latas}\033[0;40m latas de tinta, totalizando R$\033[1;35m{custo:,.2f}\033[0;40m")
print(f"Irá sobrar \033[1;36m{(latas * volumeLata) - litrosUtilizados:,.2f}\033[0;40m litros de tinta, utilizando \033[1;36m{litrosUtilizados:,.2f}\033[0;40m litros")
```

#### 14. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

Detalhe: MB significa megabyte, Mb (com b minúsculo) significa megabit. Um megabit é 1/8 de um megabyte. 

```python
tamanhoArquivo = float(input("Digite o tamanho do arquivo (em MB): "))
velocidadeInternet = float(input("Digite a velocidade da internet (em Mbps): "))
tempoEstimado = tamanhoArquivo / (velocidadeInternet / 8)

print(f"O tamanho do arquivo é de \033[1;32m{tamanhoArquivo:,.2f}\033[0;40m MB e a velocidade da internet é de \033[1;32m{velocidadeInternet:,.2f}\033[0;40m Mbps")
print(f"O tempo estimado para download é de \033[1;32m{tempoEstimado:,.2f}\033[0;40m segundos")
```
