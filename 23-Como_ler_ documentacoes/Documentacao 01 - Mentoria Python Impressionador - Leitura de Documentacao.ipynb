```python
# pandas -> analise de dados
# getting started
# quickstart
import pandas as pd

df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
})


display(df)
print(df["Age"])
# DataFrame é a representação da tabela
# pd.DataFrame -> recebe um dicionário como parâmetro e cria a tabela a partir do dicionário
# Pode receber um dicionário na estrutura do JSON
# Cada Chave representa o nome da coluna e a lista de valores representam os dados das celulas
```

```python
# datas no pandas
import pandas as pd
from datetime import datetime, timedelta

tabela = pd.read_excel('Contratos.xlsx')
tabela.info()
tabela['Data Pagamento']   = pd.to_datetime(tabela['Data Pagamento'],   format='%d/%m/%Y')
tabela['Data Pagamento 2'] = pd.to_datetime(tabela['Data Pagamento 2'], format='%m-%d-%y')
display(tabela)

comparador_data_df = tabela
# comparador_data_df['Data Pagamento 2'] = comparador_data_df['Data Pagamento 2'] + pd.to_timedelta(1, unit='D')
comparador_data_df = comparador_data_df.loc[comparador_data_df['Data Pagamento 2'] > comparador_data_df['Data Pagamento']].copy()
comparador_data_df['Diferença Data'] = comparador_data_df['Data Pagamento 2'] - comparador_data_df['Data Pagamento']
comparador_data_df.info()


display(comparador_data_df)
```

```python
# funcao merge do pandas
email = pd.read_excel('Emails.xlsx')#.rename(columns={'Nome': 'Cliente'})
#email['Nome'] = email['Nome'].rename('Cliente')
display(email)

tabela = tabela.merge(email, left_on='Cliente', right_on='Nome').drop(columns=['Nome'])
display(tabela)
```
