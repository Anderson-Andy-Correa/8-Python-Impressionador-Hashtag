#### 1. Faça um Programa que peça dois números e imprima o maior deles.

```python
from re import match
def inserirNumero(mensagem):
    while(True):
        valor = input(mensagem).strip()
        if match(r'^-?\d+(\.\d+)?$', valor):
            return float(valor)
        else:
            print("Insira um valor válido!")

qtd = 2
lista = [True for n in range(qtd)]
for posicao, item in enumerate(lista):
    lista[posicao] = inserirNumero(f"Insira o {posicao + 1}º número")
print(f"O maior número inserido foi: {max(lista)}")
```

#### 3. Faça um Programa que verifique o estado civil de uma pessoa. Se a letra digitada é "C" (Casado), "S" (Solteiro), "D" (Divorciado), "V" (Viúvo) ou "O" (outros). Conforme a letra escrita pelo usuário seu programa deve escrever o estado civil, exemplo:

Usuário digita: C

Seu programa deve responder:
C - Casado

```python
dicionario = {"C": "Casado", "S": "Solteiro", "D": "Divorciado", "V": "Viúvo", "O": "Outro"}

while(True):
    for chave, valor in dicionario.items():
        print(f'{chave} - {valor}')
    estadoUsuario = input("Digite o seu estado civil de acordo com a tabela: ").strip().upper()
    print("-" * 40)
    if estadoUsuario in dicionario.keys():
        print(f"Selecionado: {estadoUsuario} - {dicionario.get(estadoUsuario)}")
        break
    else:
        print("Opção inválida! Tente novamente.")
    print("-" * 40)
```

#### 5. Faça um programa para a leitura de duas notas parciais de um aluno. O programa deve calcular a média alcançada por aluno e apresentar:

##### A mensagem "Aprovado", se a média alcançada for maior ou igual a sete;
##### A mensagem "Reprovado", se a média for menor do que sete;
##### A mensagem "Aprovado com Distinção", se a média for igual a dez.

```python
notas = []
qtd_notas =  2
media = 0

while qtd_notas > 0:
    nota = inserirNumero(f"Insira a {len(notas) + 1}ª nota: ")
    if 0 <= nota <= 10:
        notas.append(nota)
        media += nota
    else:
        print("A nota deve estar no intervalo de 0 a 10. Tente novamente!")
        continue
    qtd_notas -= 1
media /= len(notas)
if media < 7:
    print(f"Reprovado com média {media:.1f}")
elif media < 10:
    print(f"Aprovado com média {media:.1f}")
else:
    print("Aprovado com Distinção")
```

#### 8. Faça um programa que pergunte o preço de três produtos e informe qual produto você deve comprar, sabendo que a decisão é sempre pelo mais barato.

```python
qtd_produtos = 3
lista = []
menor_preco = 0
for itens in range(qtd_produtos):
    nomeProduto = input(f"Insira o nome do {itens + 1}º produto: ")
    preco = inserirNumero(f"Insira o preço do {itens + 1}º produto: R$")
    lista.append([nomeProduto, preco])
    if itens == 0:
        menor_preco = itens
    elif lista[menor_preco][1] > preco:
        menor_preco = itens
print(f"A sugestão de compra é o produto {lista[menor_preco][0]} no valor de R${lista[menor_preco][1]}")
```
