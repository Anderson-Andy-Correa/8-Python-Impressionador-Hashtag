# Mais edições de Gráfico com MatplotLib

```python
#importando o matplotlib
import matplotlib.pyplot as plt
```

- Usando o módulo numpy para trabalhar com números

```python
#importando o módulo numpy
import numpy as np
```

### Outros tipos de Gráfico:

- Linha

```python
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)
print(vendas)
```

```python
plt.plot(meses, vendas)
plt.title('Vendas de produtos')
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.grid(True)
plt.axis([meses[0], meses[-1], 0, max(vendas) + 100])
plt.show()
```

- Editando o Gráfico de Linha

```python
# Editando a linha para apenas marcadores
def plt1():
    plt.plot(meses, vendas, '-')
    plt.title('Vendas de produtos')
    plt.xlabel('Meses')
    plt.ylabel('Vendas')
    plt.grid(True)
    plt.axis([meses[0], meses[-1], 0, max(vendas) + 100])
    

plt1()
plt.show()
```

```python
#editando o tamanho e a cor da linha - quais parâmetros posso editar?
def plt2():
    plt.plot(meses, vendas, '--o', color='purple', linewidth=2.5, markersize=8)
    plt.title('Vendas de produtos')
    plt.xlabel('Meses')
    plt.ylabel('Vendas')
    plt.grid(True)
    plt.axis([meses[0], meses[-1], 0, max(vendas) + 100])
    

plt2()
plt.show()
```

- Dispersão

```python
plt.scatter(meses, vendas)
plt.show()
```

- Barras

```python
def plt3():
    plt.bar(meses, vendas)
    
plt3()
plt.show()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
plt.figure(1, figsize=(15, 3))
plt.subplot(1, 3, 1)
plt1()
plt.subplot(1, 3, 2)
plt2()
plt.subplot(1, 3, 3)
plt3()
plt.show()
```
